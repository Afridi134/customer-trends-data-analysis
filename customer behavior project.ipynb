```python
frequency = {
    'Fortnightly': 14,
    'Bi-Weekly': 14,
    'Weekly': 7,
    'Annually': 365,
    'Quarterly': 90,
    'Every 3 Months': 90,
    'Monthly': 30
    
}

df["frequency_purchased_days"] = df["frequency_of_purchases"].map(frequency)
```

def get_frequency_days(f):
    if f in ['Fortnightly','Bi-Weekly']:
        return 14
    elif f ==  'Weekly':
        return 7
    elif f =='Annually':
        return 365
    elif f in ['Quarterly','Every 3 Months']:
        return 90
    elif f ==  'Monthly':
        return 30
    else:
        return None

df["frequency_purchased_days"] = df["frequency_of_purchases"].apply(get_frequency_days)

df.drop(columns=['frequency_purchased_days'], inplace=True)
